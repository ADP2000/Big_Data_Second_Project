```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import year, round

# Inizializza SparkSession
spark = SparkSession.builder \
    .appName("Stock Statistics") \
    .getOrCreate()
```

```python
historical_stock_prices_df = spark.read.csv("/home/addi/bigData/secondo_progetto/historical_stock_prices_restored.csv", header=True)
historical_stocks_df = spark.read.csv("/home/addi/bigData/secondo_progetto/historical_stocks.csv", header=True)
```

```python
historical_stocks_df.show()
```

```python
merged_df = historical_stock_prices_df.join(historical_stocks_df, 'ticker')
```

```python
merged_df.show()
```

```python
merged_df.createOrReplaceTempView("merged_table")
```

```python
stock_statistics_df = spark.sql("""
    SELECT ticker,
           MIN(low) as min_low                   
    FROM merged_table
    GROUP BY ticker
""")
```

```python
stock_statistics_df.show()
```

```python
stock_statistics_df.write \
    .format("csv") \
    .mode("overwrite") \
    .save("file:///./output/stock_statistics")
```

```python
merged_df = merged_df.withColumn('percent_change', round((merged_df['close'] - merged_df['open']) / merged_df['open'] * 100, 2))
```

```python
merged_df.show()
```

```python
# Raggruppa per azienda e anno, e calcola il trend di variazione annuale
trend_df = merged_df.groupBy('name', 'year').agg({'percent_change': 'avg'})
```

```python
trend_df.show()
```

```python
common_trends_df = trend_df.groupBy('name').agg({'year': 'collect_list', 'avg(percent_change)': 'collect_list'})
```

```python
common_trends_df.show()
```

```python


common_trends_df = common_trends_df.filter(common_trends_df['count(year)'] >= 3) \
    .withColumnRenamed('collect_list(year)', 'years') \
    .withColumnRenamed('collect_list(avg(percent_change))', 'trends')

# Filtra i gruppi che iniziano dal 2000
common_trends_df = common_trends_df.filter(common_trends_df['years'].contains(2000))

# Mostra i risultati
common_trends_df.show(truncate=False)

# Chiudi la sessione Spark
spark.stop()
```
