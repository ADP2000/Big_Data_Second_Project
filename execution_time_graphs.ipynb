```python
import matplotlib.pyplot as plt
```

```python

# Dati
local_time = 127.09
aws_times = [90.42, 112.79, 156.06]
file_sizes = ['Locale', 'AWS\n100%', 'AWS\n150%', 'AWS\n200%']
times = [local_time] + aws_times

# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.plot(file_sizes, times, marker='o', linestyle='-', color='b')
plt.title('Execution Time First Job - Map-Reduce')
plt.xlabel('Environment and file size')
plt.ylabel('Execution time (s)')
plt.grid(True)

# Annotazione dei valori sui punti del grafico
for i, time in enumerate(times):
    plt.text(i, time + 2, f'{time:.2f}', ha='center')

plt.show()
```
